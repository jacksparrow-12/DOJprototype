```python
import pandas as pd
df = pd.read_csv('/content/bhc101.csv', header=1)
print(df.columns)
```

```python
df.head()
```

```python
print(df.columns)
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('bert-base-nli-mean-tokens')
judge_names = df['Name of the Judge S/Shri Justice'].tolist()
judge_embeddings = model.encode(judge_names)
df['Embeddings'] = list(judge_embeddings)
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
def find_judge_info(user_input):
    input_embedding = model.encode([user_input])[0]
    similarities = cosine_similarity([input_embedding], judge_embeddings)[0]
    closest_index = np.argmax(similarities)
    judge_info = df.iloc[closest_index]
    return judge_info
user_input = "DEVENDRA KUMAR UPADHYAYA"
judge_info = find_judge_info(user_input)
print(judge_info)
```

```python
def chatbot():
    print("Welcome to the DoJ chatbot. Ask about any judge by name.")
    while True:
        user_input = input("Enter judge's name (or type 'exit' to quit): ")
        if user_input.lower() == 'exit':
            print("Goodbye!")
            break
        judge_info = find_judge_info(user_input)
        if judge_info is not None:
            print(f"Judge Information:\n{judge_info}\n")
        else:
            print("Sorry, I couldn't find any information on that judge.")
chatbot()
```
